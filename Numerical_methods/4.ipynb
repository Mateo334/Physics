# Úloha 4: Obvod Rocheova laloku v ekvatoriální rovině

*Numerické řešení rovnic, analýza chyby, extrapolace*

V popisu dynamiky dvojhvězdných systémů je definován **Rocheův lalok** jako oblast prostoru kolem hvězdy 1, kterou lze zaplnit plynem, aniž by tento plyn přetékal na hvězdu 2. Tato oblast je ohraničena **ekvipotenciální plochou** v systému korotujícím se vzájemně se obíhajícími hvězdami, na níž je potenciál roven potenciálu v **Lagrangeově bodě $L_1$** (bod, kde se vyrovnávají gravitační síly obou těles a odstředivá síla).

V této úloze se omezíme na rovinu, ve které obíhají obě hvězdy po kruhové dráze. V této rovině nalezněte čáru ohraničující Rocheův lalok a spočtěte její délku. Počátek souřadného systému položte do hvězdy 1 a osa $x$ nechť směřuje ke hvězdě 2. Délky budeme měřit v jednotkách daných vzdáleností obou hvězd $R$.

## Podrobnější instrukce

- Najděte polohu bodu $L_1$ a hodnotu potenciálu $V_L$ v tomto bodě.
- Pro hodnoty  
  $$
  \theta = \frac{2\pi n}{N}, \qquad n = 0,1,\dots,N
  $$
  nalezněte body  
  $$
  x_n = r \cos\theta, \qquad y_n = r \sin\theta
  $$
  ležící na hledané ekvipotenciální čáře. Pro každé pevné $\theta$ hledejte $r$ řešením rovnice  
  $$
  V(r) = V_L
  $$
  vámi zvolenou numerickou metodou. Pokuste se dosáhnout přesnosti blízké strojovému $\epsilon$.


- Spočtěte délku $D_N$ lomené čáry  
  $$
  (x_1, y_1), (x_2, y_2), \ldots, (x_N, y_N).
  $$
  Studujte konvergenci této hodnoty pro $N \to \infty$ a pokuste se identifikovat bod, kde dochází k rovnováze mezi zaokrouhlovací a diskretizační chybou.
- Pokuste se přesnost metody zlepšit, například vhodnou extrapolací diskretizační chyby.

## Výstup

Výstupem k odevzdání bude:
- obrázek lomené čáry  
  $$
  (x_1, y_1), (x_2, y_2), \ldots, (x_N, y_N),
  $$
- obrázek dokumentující konvergenci $D_N$, ze kterého bude patrný řád diskretizační chyby,
- co nejpřesnější hodnota  
  $$
  D = \lim_{N \to \infty} D_N.
  $$

## Rovnice

### Poloha Lagrangeova bodu $L_1$

Souřadnice $x_L$ bodu $L_1$ je dána rovnováhou sil:
$$
x_L - m_2 + \frac{m_2}{(1 - x_L)^2} - \frac{m_1}{x_L^2} = 0.
$$

Zde platí:
$$
m_1 = \frac{M_1}{M}, \qquad
m_2 = \frac{M_2}{M}, \qquad
M = M_1 + M_2.
$$

### Gravitační potenciál

Gravitační potenciál v korotujícím systému (včetně odstředivé síly), vyjádřený v jednotkách $\kappa M / R$, má tvar:
$$
V(x, y) = -\frac{m_1}{r} - \frac{m_2}{r_2} - \frac{1}{2} r_0^2,
$$
kde
$$
\begin{aligned}
r &= \sqrt{x^2 + y^2}, \\
r_2 &= \sqrt{r^2 - 2r\cos\theta + 1}, \\
r_0 &= \sqrt{r^2 - 2m_2 r\cos\theta + m_2^2}.
\end{aligned}
$$
