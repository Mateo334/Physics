# Úloha 5: Model oscilující chemické reakce

*Řešení obyčejných diferenciálních rovnic*

V dnešní době je známo několik tzv. **chemických oscilátorů** (doporučuji vyhledat „chemical oscillator“ na YouTube). Jedná se o směsi látek, v nichž dochází k soustavám chemických reakcí vedoucích k oscilacím koncentrace meziproduktů. V době svého objevu toto chování vedlo k nedůvěře a odmítnutí (viz *Belousov–Zhabotinsky reaction* na anglické Wikipedii), dnes však existují jednoduché teoretické modely, které tyto jevy vysvětlují.

Jeden takový jednoduchý model (Lotka–Volterra model) si zde prozkoumáme. V tomto modelu se látka **A** postupně mění na látku **B**, přičemž vznikají meziprodukty **X** a **Y** podle schématu

$$
\mathrm{A} + \mathrm{X} \rightarrow 2\mathrm{X}
$$

$$
\mathrm{X} + \mathrm{Y} \rightarrow 2\mathrm{Y}
$$

$$
\mathrm{Y} \rightarrow \mathrm{B}
$$

Označíme-li $k_1$, $k_2$ a $k_3$ reakční rychlosti těchto reakcí, dostaneme pro časové závislosti koncentrací $A$, $X$, $Y$ a $B$ následující soustavu diferenciálních rovnic:

$$
\begin{array}{l}
\frac{d}{dt} A = -k_1 A X, \qquad
\frac{d}{d\tau} a = -\kappa a x, \\[6pt]
\frac{d}{dt} X = k_1 A X - k_2 X Y, \qquad
\frac{d}{d\tau} x = \kappa a x - x y, \\[6pt]
\frac{d}{dt} Y = k_2 X Y - k_3 Y, \qquad
\frac{d}{d\tau} y = x y - y, \\[6pt]
\frac{d}{dt} B = k_3 Y, \qquad
\frac{d}{d\tau} b = y.
\end{array}
$$

V pravém sloupci jsou rovnice přepsány do bezrozměrných veličin

$$
\tau = t k_3, \qquad
b = \frac{B}{A_0}, \qquad
y = \frac{Y}{A_0}, \qquad
x = \frac{k_2 X}{k_3}, \qquad
\kappa = \frac{k_1}{k_2},
$$

kde $A_0$ je počáteční koncentrace látky **A**.

Řešte tyto rovnice numericky pro $\kappa = 0.002$ a počáteční podmínky

$$
a = 498.4, \qquad
x = 1.5, \qquad
y = 0.1, \qquad
b = 0.
$$

## Úkoly

- Napište proceduru pro integraci dané soustavy diferenciálních rovnic s řádem přesnosti 4. řádu.  
  Ověřte konvergenci výsledků v závislosti na délce kroku pro výsledek ve fixním čase $t = 100$ a dále ověřte, že veličina  
  $$
  a + x + y + b
  $$
  se zachovává.
- Nakreslete $a(t)$ a $b(t)$ do jednoho obrázku a nalezněte čas $t$, kdy se vyrovná koncentrace reaktantů a produktů,
  $$
  a(t) = b(t).
  $$
  Odhadněte, s jakou přesností jste tento čas určili.
- Nakreslete koncentrace meziproduktů $x(t)$ a $y(t)$ a určete periodu jejich oscilací. Odhadněte, jaká je přibližná přesnost určení této periody.

Rozmyslete si, že oscilace fungují jako jakási **chemická pumpa**: část cyklu s vysokým $x$ vede ke zvýšené spotřebě reaktantu **A**, zatímco část cyklu s vysokým $y$ vede ke zvýšené tvorbě produktu **B**.

## Výstup

Výstupem budou:
- tři obrázky (rychlost konvergence, časová závislost $a(t)$ a $b(t)$, časová závislost meziproduktů),
- dvě číselné hodnoty (čas vyrovnání koncentrací a perioda oscilací).
